# Lambda

```python
def double(x):
    return x*2

print(double(5))    
```

```python
double = lambda x : x*2
print(double(5))
```

```python
cube = lambda x: x*x*x
print(cube(3))
```

```python
def cube(x):
    return x*x*x
print(cube(3))    
```

```python
def appl(fx, value):  #func ko func pass kar sakte ho
    return 6 + fx(value)
print(appl(cube, 2))    
```

# Map, Filter & Reduce

### Map

```python
def cube(x):
    return x*x*x

print(cube(6))    
```

```python
l = [1,2,3,4,4]
```

```python
for i in l:
    n = cube(i)
    print(n)
```

```python
new_l = list(map(cube, l)) #tuple, set can also be used instead of list
print(new_l)
```

### Filter

```python
def filter_func(a):
    return a<4


newnewl = list(filter(filter_func, l))
print(newnewl)
```

### Reduce

```python
from functools import reduce
```

```python
numbers = [1,2,3,4,5,6]
```

```python
sum = reduce(lambda x, y : x+y, numbers)
print(sum)
```

```python
def mysum(x,y):
    return x+y
```

```python
s = reduce(mysum, numbers)
print(s)
```

# Snake, Water, Gun

```python
import random
l = ["snake", "water", "gun"]
n1 = random.choice(l)
print("\033[1mWELCOME TO THE GAME!!\033[0m")
n = input("You :")
n = n.lower()
print("Computerji:", n1)

if n not in l:
    print("Enter valid input!")
else:
    if n == n1:
        print("It's a tie!")
    elif (n == "snake" and n1 == "water") or (n == "water" and n1 == "gun") or (n == "gun" and n1 == "snake"):
        print("You win!!")
    else:
        print("Computer wins!")
```
